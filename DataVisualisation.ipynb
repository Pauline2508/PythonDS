```python
import matplotlib.pyplot as plt



# Sample data

x = [1, 2, 3, 4, 5]

y = [10, 8, 12, 6, 20]



# Create a line plot

plt.plot(x, y)



# Add labels and title

plt.xlabel('X-axis label')

plt.ylabel('Y-axis label')

plt.title('Line Plot Example')



# Display the plot

plt.show()
```

```python
import matplotlib.pyplot as plt



# Sample data

categories = ['Category A', 'Category B', 'Category C', 'Category D']

values = [30, 20, 15, 35]



# Create a pie chart using Matplotlib

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)



# Add title

plt.title('Pie Chart Example')



# Display the plot

plt.show()
```

```python
import plotly.graph_objects as go



# Sample data

x = [1, 2, 3, 4, 5]

y = [10, 8, 12, 6, 14]

z = [5, 9, 3, 7, 2]



# Create a 3D scatter plot using Plotly

fig = go.Figure(data=go.Scatter3d(x=x, y=y, z=z, mode='markers'))



# Add title

fig.update_layout(title='3D Scatter Plot Example')



# Show the plot

fig.show()
```

```python
import matplotlib.pyplot as plt



# Sample data

years = [2010, 2011, 2012, 2013, 2014]

series_a = [20, 25, 30, 35, 40]

series_b = [15, 18, 25, 28, 30]



# Create an area plot using Matplotlib

plt.stackplot(years, series_a, series_b, labels=['Series A', 'Series B'], alpha=0.6)



# Add labels and title

plt.xlabel('Years')

plt.ylabel('Values')

plt.title('Area Plot Example')

plt.legend(loc='upper left')



# Display the plot

plt.show()
```

```python
import seaborn as sns

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt



# Sample data - create a pandas DataFrame

data = {

    'A': np.random.randn(100),

    'B': np.random.randn(100),

    'C': np.random.randn(100),

    'D': np.random.randn(100)

}



df = pd.DataFrame(data)



# Calculate the correlation matrix

correlation_matrix = df.corr()



# Create a heatmap using Seaborn

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)



# Add title

plt.title('Correlation Plot Heatmap')



# Display the plot

plt.show()
```

```python
import pandas as pd

data = {
    'Region': ['East', 'West', 'East'],
    'Product': ['A', 'B', 'C'],
    'Sales': [100, 150, 200]
}

sales_data = pd.DataFrame(data)

total_sales_by_region = sales_data.groupby('Region')['Sales'].sum().reset_index()

print(total_sales_by_region)
```

```python
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [25, None, 28, None, 22],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female']
}

# Create the DataFrame
df = pd.DataFrame(data)

# Calculate the mean age
mean_age = df['Age'].mean()

# Fill missing values in 'Age' column with the mean age
df['Age'].fillna(mean_age, inplace=True)

print(df)
```

```python
import pandas as pd

data = {
    'Employee_ID': [101, 102, 101, 103, 102],
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Salary': [50000, 60000, 50000, 55000, 60000]
}

employee_data = pd.DataFrame(data)

employee_data.drop_duplicates(subset=['Employee_ID'], keep='first', inplace=True)

print(employee_data.reset_index())
```
